```python
%%writefile market_risk_platform/data_ingestion/fetch_nikkei.py
import yfinance as yf
from utils.logger import get_logger

logger = get_logger(__name__)

def fetch_nikkei_data(ticker, start, end):
    logger.info(f"Fetching Nikkei data: {ticker}")
    df = yf.download(ticker, start=start, end=end)

    if df.empty:
        raise ValueError("No market data fetched")

    df.reset_index(inplace=True)
    return df
```

```python
%%writefile market_risk_platform/data_ingestion/validate_nikkei.py
def validate_nikkei_data(df):
    required_columns = ["Date", "Open", "High", "Low", "Close"]

    for col in required_columns:
        if col not in df.columns:
            raise ValueError(f"Missing column: {col}")

    if not df["Date"].is_monotonic_increasing:
        raise ValueError("Dates are not sorted")

    if (df[["Open", "High", "Low", "Close"]] <= 0).any().any():
        raise ValueError("Invalid non-positive prices")

    if df.isnull().mean().max() > 0.01:
        raise ValueError("Too many missing values")

    return df
```

Feature engineering

```python
%%writefile market_risk_platform/features/returns.py
import numpy as np

def log_returns(df, price_col="Close"):
    df["log_return"] = np.log(df[price_col] / df[price_col].shift(1))
    return df
```

```python
%%writefile market_risk_platform/features/volatility.py
def rolling_volatility(df, window=20):
    df[f"realized_vol_{window}"] = df["log_return"].rolling(window).std()
    return df
```

```python
%%writefile market_risk_platform/features/lags.py
def lag_features(df, lags=[1, 2, 5, 10]):
    for lag in lags:
        df[f"lag_return_{lag}"] = df["log_return"].shift(lag)
    return df
```

```python
%%writefile market_risk_platform/features/build_features.py
from features.returns import log_returns
from features.volatility import rolling_volatility
from features.lags import lag_features

def build_features(df):
    df = log_returns(df)
    df = rolling_volatility(df, window=20)
    df = lag_features(df)
    return df.dropna()
```

```python
%%writefile market_risk_platform/models/garch_models.py
from arch import arch_model
import pandas as pd

def fit_garch(returns):
    model = arch_model(
        returns * 100,
        vol="Garch",
        p=1,
        q=1,
        dist="normal"
    )
    return model.fit(disp="off")

def fit_egarch(returns):
    model = arch_model(
        returns * 100,
        vol="EGarch",
        p=1,
        q=1,
        dist="normal"
    )
    return model.fit(disp="off")

def extract_volatility(result, index):
    return pd.Series(
        result.conditional_volatility / 100,
        index=index
    )
```

```python
import numpy as np

def qlike_loss(realized_vol, model_vol):
    return np.mean(
        np.log(model_vol**2) + (realized_vol**2) / (model_vol**2)
    )
```

```python
qlike_garch = qlike_loss(
    eval_df["realized_vol_20"].values,
    eval_df["garch_vol"].values
)

qlike_egarch = qlike_loss(
    eval_df["realized_vol_20"].values,
    eval_df["egarch_vol"].values
)

print("QLIKE GARCH :", qlike_garch)
print("QLIKE EGARCH:", qlike_egarch)
```

Value at Risk (VaR) & CVaR + Backtesting

```python
returns = df_feat["log_return"].dropna()
vol = df_feat.loc[returns.index, "garch_vol"]
```

```python
from scipy.stats import norm

def parametric_var(vol, alpha):
    z = norm.ppf(alpha)
    return z * vol
```

```python
var_95 = parametric_var(vol, 0.05)
var_99 = parametric_var(vol, 0.01)
```

```python
def cvar(returns, var_series):
    return returns[returns <= var_series].mean()
```

```python
cvar_95 = returns.rolling(250).apply(
    lambda x: x[x <= np.quantile(x, 0.05)].mean(),
    raw=False
)
```

```python
def var_breaches(returns, var_series):
    return returns < var_series
```

```python
breach_95 = var_breaches(returns, var_95)
breach_99 = var_breaches(returns, var_99)

print("95% breach rate:", breach_95.mean())
print("99% breach rate:", breach_99.mean())
```

```python
%%writefile market_risk_platform/models/ml_baselines.py
from sklearn.linear_model import LinearRegression, Ridge, ElasticNet

def fit_linear(X, y):
    model = LinearRegression()
    model.fit(X, y)
    return model

def fit_ridge(X, y, alpha=1.0):
    model = Ridge(alpha=alpha)
    model.fit(X, y)
    return model

def fit_elasticnet(X, y, alpha=0.01, l1_ratio=0.5):
    model = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)
    model.fit(X, y)
    return model
```

```python
%%writefile market_risk_platform/models/ml_tree_models.py
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

def fit_random_forest(X, y, n_estimators=200, max_depth=5):
    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        random_state=42
    )
    model.fit(X, y)
    return model

def fit_xgboost(X, y):
    model = XGBRegressor(
        n_estimators=300,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="reg:squarederror",
        random_state=42
    )
    model.fit(X, y)
    return model
```

```python
%%writefile market_risk_platform/models/ml_evaluation.py
import numpy as np

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def qlike(y_true, y_pred):
    eps = 1e-8
    y_pred = np.maximum(y_pred, eps)
    return np.mean(np.log(y_pred) + (y_true / y_pred))
```

```python
%%writefile market_risk_platform/pipelines/train_ml_models.py
from models.ml_baselines import fit_linear, fit_ridge, fit_elasticnet
from models.ml_tree_models import fit_random_forest, fit_xgboost
from models.ml_evaluation import mse, qlike

def train_and_evaluate(X_train, X_test, y_train, y_test):
    models = {
        "Linear": fit_linear(X_train, y_train),
        "Ridge": fit_ridge(X_train, y_train),
        "ElasticNet": fit_elasticnet(X_train, y_train),
        "RandomForest": fit_random_forest(X_train, y_train),
        "XGBoost": fit_xgboost(X_train, y_train)
    }

    results = {}

    for name, model in models.items():
        preds = model.predict(X_test)
        results[name] = {
            "MSE": mse(y_test, preds),
            "QLIKE": qlike(y_test, preds)
        }

    return results
```
